```python
import numpy as np 
import matplotlib.pyplot as plt 
```

Nous avons utilisé la solution explicite de I(t) trouvée aux pages 12 et 13 du PDF. Explicite_SIS correspond 
aux cas $R_0 < 1$ et $R_0 >1$. Explicite_SIS_0 correspond au cas $R_0 = 1$. On considère que la proportion
initiale d'infectés est de 30%. 


```python
def Explicite_SIS(t,a,b):
    I0 = 0.3
    lam = 1/I0 - b/(b-a) 
    return (1-1/(lam*np.exp((a-b)*t)+b/(b-a)), 1/(lam*np.exp((a-b)*t)+b/(b-a)))

def Explicite_SIS_0(t,b):
    I0 = 0.3
    return (1-I0/(b*t*I0 + 1), I0/(b*t*I0 + 1))
    
```

```python
t = np.linspace(0,100,1000)
plt.figure(figsize=(15,5))
A = [ r'Cas $R_0 = 1$ pour $\beta= \frac{1}{2}$' , r'Cas $R_0 < 1$ pour $\alpha = \frac{1}{2}$ et $\beta=\frac{2}{5} $', r'Cas $R_0 > 1$ pour $\alpha = \frac{1}{6}$ et $\beta= \frac{1}{2}$']
Solutions = [Explicite_SIS_0(t,1/2), Explicite_SIS(t,1/2,2/5), Explicite_SIS(t,1/6,1/2)]
for i in range(3):
    plt.subplot(1,3,i+1)
    plt.plot(t,Solutions[i][1], label='I(t)')
    plt.plot(t,Solutions[i][0], label='S(t)')   
    if i == 2:
        plt.text(30, 0.1, r'$ \lim_{t \rightarrow + \infty}I(t) = 1-\frac{\alpha}{\beta} = \frac{2}{3} $', size = 'x-large'  )
    else : 
        plt.text(60,0.1, r'$\lim_{t \rightarrow + \infty}I(t) = 0$', size = 'x-large' )
    plt.ylim(0,1)
    plt.xlabel("Temps t")
    plt.ylabel("Concentrations S et I") 
    plt.title(A[i])
    plt.legend()
```
